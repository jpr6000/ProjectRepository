  This ER diagram represents the relationships between entities that exist in the Uber business model. The ER diagram includes 5 entities: The Company, Drivers, Customers, the Uber App, and Locations. In the diagram, the company is given a company ID as the primary key to identify it and the company's state, zip code, and street number are all attributes of the entity itself. Additionally, Drivers are employees of the company, resulting in the drivers foreign key to be present in the Company table. There is only one Uber company, which employs anywhere from one driver to three million drivers. The Driver table includes the driver ID as the primary key, as each driver is given a unique ID. The Driver table also includes the drivers first name, last name, and their date of birth. A foreign key is also present in the Driver table which is the customer that they drive, linking the Driver and Customer tables. Only one driver drives anywhere from one to four customers. The Customer table's primary key is the customer's ID as each customer is given a unique ID by Uber. The Customer table also includes the customer's first name, last name, and date of birth as regular attributes. The Customer table has a location foreign key to identify the location they would like to go to and a Uber app profile foreign key which connects the customer to their app profile. One to four customers can go to one and only one location per driving session. The location table has a primary key of a location ID and attributes of the zip code, the state, and the street number. One to four customer can use the Uber app per drive session. The Uber app table has an app ID as the primary key to identify the app on the app store. The Uber app table has attributes that describe functions of the app including the rating system, the ride scheduler, and for transactions.
